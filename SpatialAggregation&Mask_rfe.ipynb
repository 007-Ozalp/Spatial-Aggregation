In this section the <a href="https://regionmask.readthedocs.io/en/v0.5.0/notebooks/mask_xarray.html">regionmask</a> package is  used to create masks of geographic regions for arbitrary longitude and latitude grids inside the function <font color='red'> ****Mask_Area_Plot**** </font>. Hence, at first step the function:

1. reads the tiff files in <a href="https://docs.xarray.dev/en/stable/">xarray</a> to compute the arithmetic mean along the labeled coordinates to define them for the <font color='teal'> **regionmask**.

2. then read the csv format file in <a href="https://pandas.pydata.org/docs/getting_started/index.html#getting-started">pandas</a>  **clip** </font>the tiff files by 
<a href="https://pygeos.readthedocs.io/en/stable/installation.html">pygeos</a> geometry.

3. Finally, save file. **

```python
file_roi= pd.read_csv('AFRICA_MASK.csv')
```

## <a id='section5'></a>5. Plot
[Back to top](#TOC_TOP)

</div>

```python
read_mask_1=xarray.open_dataset(data_resampled_mask_1)
read_mask_2=xarray.open_dataset(data_resampled_mask_2)
read_mask_3=xarray.open_dataset(data_resampled_mask_3)
read_mask_4=xarray.open_dataset(data_resampled_mask_4)
```

```python
read_mask_1#,read_mask_2,read_mask_3,read_mask_4
```

```python
plt.rcParams["figure.figsize"] = [16, 8]
plt.rcParams["figure.autolayout"] = False

fig, (ax1, ax2) = plt.subplots(ncols=2,sharex=True,sharey=True,gridspec_kw={'width_ratios': [0.8, 1]})

fig.add_subplot(111, frameon=False)

# hide tick and tick label of the big axis
plt.tick_params(labelcolor='none', which='both', top=False, bottom=False, left=False, right=False)


plot1=read_mask_1.band_data.plot(ax=ax1,cmap='terrain_r',vmin=0,vmax=250,add_colorbar=False)

plot2=read_mask_2.band_data.plot(ax=ax2,cmap='terrain_r',vmin=0,vmax=250,add_colorbar=False)

fig.suptitle('RFE',x=0.49,fontsize = 20,weight='bold')
ax1.set_title('TAMSAT',fontsize = 16)
ax1.set_ylabel('')
ax1.set_xlabel('')

ax2.set_title('CHIRP',fontsize = 16)
ax2.set_ylabel('')
ax2.set_xlabel('')


plt.xlabel("\nLONGITUDE",x=0.45,size=18)
plt.ylabel("LATITUDE",size=18)

cbar2 = plt.colorbar(plot2,ax=ax2)
cbar2.set_label('\nRFE[mm]',labelpad=+38, rotation=270,size=12)

plt.autoscale()
plt.tight_layout()

plt.show()
```
